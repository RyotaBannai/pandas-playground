```python
import pandas.util.testing as tm; tm.N = 3
def unpivot(frame):
    N, K = frame.shape ##3,4
    data = {'value' : frame.values.ravel('F'),
            'variable' : np.asarray(frame.columns).repeat(N),
            'date' : np.tile(np.asarray(frame.index), K)}
    return pd.DataFrame(data, columns=['date', 'variable', 'value'])
df = unpivot(tm.makeTimeDataFrame());
```

```python
import datetime
df = pd.DataFrame({'A': ['one', 'one', 'two', 'three'] * 6,
                   'B': ['A', 'B', 'C'] * 8,
                   'C': ['foo', 'foo', 'foo', 'bar', 'bar', 'bar'] * 4,
                   'D': np.random.randn(24),
                   'E': np.random.randn(24),
                   'F': [datetime.datetime(2013, i, 1) for i in range(1, 13)] +
                   [datetime.datetime(2013, i, 15) for i in range(1, 13)]})
```

```python
pd.pivot_table(df, values='D', index=['A', 'B'], columns=['C'])
```

```python
pd.pivot_table(df, values='D', index=['B'], columns=['A', 'C'], aggfunc=np.sum)
```

```python
#you can use Grouper for index
t = pd.pivot_table(df, values='D', index=pd.Grouper(freq='M', key='F'), columns='C');t
```

```python
print(t.to_string(na_rep=''))
```

```python
df.pivot_table(index=['A', 'B'], columns='C', margins=True, aggfunc=("mean"))
```

```python
##if the same data are more than two, return the mean values
##change the aggfunc if you want others values form, such as sum, np.std
df.pivot_table(index=['A', 'B'], columns='C', aggfunc="sum")
```

```python
foo, bar, dull, shiny, one, two = 'foo', 'bar', 'dull', 'shiny', 'one', 'two'
a = np.array([foo, foo, bar, bar, foo, foo], dtype=object)
b = np.array([one, one, two, one, two, one], dtype=object)
c = np.array([dull, dull, shiny, dull, dull, shiny], dtype=object)
pd.crosstab(a, [b, c], rownames=['a'], colnames=['b', 'c'])
```
